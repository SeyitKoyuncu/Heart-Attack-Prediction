### Bootstrap Sampling

```python
import numpy as np
import random

x = np.random.normal(loc= 300.0, size=1000)
print(f"True mean is = {np.mean(x)}")

sample_mean = []
for i in range(3):
  y = random.sample(x.tolist(), 4)
  avg = np.mean(y)
  sample_mean.append(avg)

print(f"With bootstraping mean is = {np.mean(sample_mean)}")
print("With these mean results we can infer our data is not have so much outlier because both meanings is near")
```

# Feature Selection

### Chi2 

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

'''
score_func = Function taking two arrays X and y, and returning a pair of arrays (scores, pvalues) or a single array with scores. 
Default is f_classif (see below “See Also”). The default function only works with classification tasks.

k = Number of top features to select. The “all” option bypasses selection, for use in a parameter search.
'''

KBest = SelectKBest(score_func = chi2, k = 12)
fit = KBest.fit(x, y)
fit_scores = fit.scores_

features = fit.transform(x)
print(fit_scores)
```

### Mutual Information

```python
from sklearn.feature_selection import mutual_info_classif
KBest2 = SelectKBest(score_func = mutual_info_classif, k = 10)

fit2 = KBest2.fit(x,y)
fit_scores2 = fit2.scores_

features2 = fit2.transform(x)
print(fit_scores2)
```

### Clear The Age

```python
df_v2 = df.drop('age', 1) #cleaning the age column since not distrubited equal, low correleation between output and have very low mutual information value. 
#Maybe will be added in the wrapper models.
df_v2
```

### Binning in the Age Feature

```python
df_binning=df
age_29_50=[]
age_50_77=[]
for item in df["age"]:
    if 50>item>=29:
        age_29_50.append(1)
    else:
        age_29_50.append(0)
    
    if 50<=item<=77:
        age_50_77.append(1)
    else:
        age_50_77.append(0)

df_binning["age_29_50"]=age_29_50
df_binning["age_50_77"]=age_50_77



df_binning["output_heart_dis"]=df_binning["output"]
df_binning=df_binning.drop("output",axis=1)
df_binning=df_binning.drop("age",axis=1)
df_binning.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
x_train_v2, x_test_v2, y_train_v2, y_test_v2 = train_test_split(x_new, y_new, test_size=0.25, random_state=0)


from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

#scale original data
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)

#scale without age feature
x_train_sc_v2 = sc.fit_transform(x_train_v2)
x_test_sc_v2 = sc.transform(x_test_v2)


#scale with robust
from sklearn import preprocessing
scaler = preprocessing.RobustScaler()
x_train_robust = scaler.fit_transform(x_train)
x_test_robust = scaler.fit_transform(x_test)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

#original data
rnd = RandomForestClassifier()
rnd.fit(x_train,y_train)

#prediction
y_preds = rnd.predict(x_test)

#score of train set
score_train=rnd.score(x_train,y_train)
print(score_train)

#score of test set
score_test=rnd.score(x_test,y_test)
print(score_test)

accuracy_rnd = accuracy_score(y_preds, y_test)
```

```python
conf_mat = confusion_matrix(y_test, y_preds)
print(conf_mat)

sns.heatmap(conf_mat,annot=True)
print(accuracy_score(y_test, y_preds))
print(confusion_matrix(y_test, y_preds))
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_params = {"cosine", "euclidean", "manhattan", "nan_euclidean"}
best_accuracy_knn = 0
best_metric = ""
best_neighbor_count = 0

for i in range(1,10):
    for param in knn_params:
        knn = KNeighborsClassifier(n_neighbors=i, metric = param, weights = "distance")
        knn.fit(x_train, y_train)
        y_pred = knn.predict(x_test)
        if best_accuracy_knn < accuracy_score(y_test, y_pred):
            best_accuracy_knn, best_neighbor_count, best_metric = accuracy_score(y_test, y_pred), i, param

print(f"Best accuracy with no scaler is = {best_accuracy_knn}, with {best_neighbor_count} n_neighbors and {best_metric} metric")
knn = KNeighborsClassifier(n_neighbors=best_neighbor_count, metric = best_metric, weights = "distance")
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
print(confusion_matrix(y_test,y_pred))
```

```python
print(f"Accuracy score with age  = {accuracy_score(y_test, y_preds)}")
print(confusion_matrix(y_test, y_pred))
```

### SVM

### Roc Curve For Random Forest

```python
from sklearn.metrics import roc_auc_score, roc_curve

y_pro=rnd.predict_proba(x_test)

y_pro_pos=y_pro[:,1]
print(y_pro_pos)

import matplotlib.pyplot as plt

def visual_roc_curve(y_test, y_pro_pos):

    fpr, tpr, thresholds = roc_curve(y_test, y_pro_pos)
    plt.plot(fpr, tpr,color="red",label="ROC")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.plot([0,1],[0,1],color="blue")
    plt.title("ROC with age")
  

visual_roc_curve(y_test,y_pro_pos)
```
